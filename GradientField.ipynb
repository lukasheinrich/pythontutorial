```python
import sympy as sy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x,y = sy.symbols('x y')

f = -(x**2 + 2*y**2)

ff = sy.lambdify((x,y),f)
dfdx = sy.lambdify((x,y),f.diff(x))
dfdy = sy.lambdify((x,y),f.diff(y))
```

```python
grid = x,y = np.mgrid[-5:5:11j,-5:5:21j]
points = np.moveaxis(grid,0,-1).reshape(11*21,2)


v=ff(points[:,0],points[:,1])
vdfdx=dfdx(points[:,0],points[:,1])
vdfdy=dfdy(points[:,0],points[:,1])

plt.contourf(x,y,v.reshape(11,21),levels = 100)
plt.contour(x,y,v.reshape(11,21),levels = 10, colors = 'k')
plt.quiver(x,y,vdfdx,vdfdy, color = 'w')
plt.gcf().set_size_inches(5,5)
```
