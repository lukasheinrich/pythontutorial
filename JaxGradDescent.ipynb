```python
import jax
import jax.numpy as jnp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def func(x):
    return x**2
```

```python
x = np.linspace(-10,10)
```

```python
w = 10.0
for i in range(1000):
    a = func(w)
    if i % 5 == 0:
        plt.plot(x,func(x))
        plt.scatter(w,a)
        plt.show()
    w-=0.001*jax.grad(func)(w)
```
