```python
import math
def func(x,y):
    a = x*y
    b = math.sin(x)
    z = a+b
    return z


def shadow_func(point,diff):
    x,y = point
    dx,dy = diff 
    da = x * dy + y * dx
    db = math.cos(dx) * dx
    dz = da + db
    return dz

func(1,2),shadow_func((1,2),(1,0)),shadow_func((1,2),(0,1))
```
